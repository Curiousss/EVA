```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

Kernel for vertical edge detection has a central vertical axis of 0 and a positive and negative values on either sides. These opposite values create a contrast in the input image to highlight any exisitng edges.


```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

Another Kernel for vertical edge detection, just detects the edge from the other direction. It has a central vertical axis of 0 and a positive and negative values on either sides. These opposite values create a contrast in the input image to highlight any exisitng edges.


```
# OTHER VERTIAL FILTER
kernel = np.float32([[1,0,-1],[1,0,-1],[1,0,-1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

Kernel for right angle detection has a central diagonal axis of 0 and a positive and negative values on either sides of the corner. These opposite values create a contrast in the input image to highlight any exisitng right angles.

The Blur filter reduce the intensity of the input image in a circular gradient. That is, the intensity of the pixels at the edges are lowest and the intensity of the centre is the highest. The kernel has center value that slightly reduces the intensity of the centre pixel of the input receptive field by 1/4. The immediate values around the centre of the kernel reduce the intensity further to 1/8 for the 4 respective pixels in the receptive field. The farthest reduce to 1/8 for respective pixels.


```
#BLUR FILTER
kernel = np.float32([[1/16, 1/8, 1/16],[1/8,1/4,1/8],[1/16,1/8,1/16]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

The shrapen filter highlights the centre while blurs the surrounding pixels of the centre to create intense contrast resulting in sharpening of an image.
The kernel has a centre value of 5 which intensifies the centre and -1 in the surround pixels which blurs the surrounding pixels to highlight the centre further.

```
#SHARPEN FILTER
kernel = np.float32([[0, -1, 0],[-1,5,-1],[0,-1,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

The identity kernel output the input image as it is. Mathematically the value one in the centre extracts the respective pixels from the input image as it is.

```
#IDENTITY FILTER
kernel = np.float32([[0,0,0],[0,1,0],[0,0,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```
